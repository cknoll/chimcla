```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming img is the grayscale image and mask is the binary array acting as the mask
img = np.random.randint(0, 256, (20, 100))  # Example grayscale image
mask = np.random.randint(0, 2, (20, 100))  # Example binary mask

# Display the grayscale image
plt.imshow(img, cmap='gray')

# Draw the outline of the mask
plt.contour(mask, levels=[0.5], colors='red', linewidths=2)

# Show the plot
plt.show()
```
